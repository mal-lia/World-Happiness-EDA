Meskipun ada negara yang memiliki skor di atas 7, jumlahnya tidak banyak.** Distribusi ini menggambarkan bahwa secara global, *mayoritas negara belum mencapai tingkat kebahagiaan yang sangat tinggi, namun cenderung berada di tingkat menengah.*

# **Handling Outliers (Menangani Nilai Pencilan)**

```python
# Membuat Boxplot
plt.figure(figsize=(8, 5))
ax = sns.boxplot(
    x=df['Happiness Score'],
    color="lightblue",
    width=0.5,
    fliersize=5  # Ukuran Titik Outlier
)

# Menambahkan Garis Mean
mean_val = df['Happiness Score'].mean()
plt.axvline(mean_val, color='navy', linestyle='--', label=f'Mean: {mean_val:.2f}')

# Tambahan Informasi Visual
plt.title('Distribusi Skor Kebahagiaan (Happiness Score)', fontsize=14)
plt.xlabel('Happiness Score', fontsize=12)
plt.legend()

# Tampilkan Plot
plt.tight_layout()
plt.show()
```

> **Analisis Boxplot Distribusi Skor Kebahagiaan**

* Visualisasi menunjukkan distribusi dari variabel "Happiness Score". Dalam boxplot ini, kotak utama mewakili **rentang nilai kuartil pertama (Q1) hingga kuartil ketiga (Q3), yaitu 50% nilai tengah dari data. Garis vertikal di dalam kotak menunjukkan median (nilai tengah dari keseluruhan data), yang berada di sekitar nilai 5,5.** Sumbu horizontal (x-axis) diberi label "Happiness Score", yang berarti bahwa variabel ini diukur pada skala skor kebahagiaan.

* Ekstensi garis dari kotak (disebut whisker) menunjukkan **rentang nilai minimum dan maksimum yang masih dianggap tidak ekstrem (tanpa outlier).** Berdasarkan plot ini, distribusi data terlihat **relatif simetris, karena median berada cukup di tengah kotak dan whisker kiri-kanan tampak seimbang.** Selain itu, tidak tampak titik-titik di luar whisker, yang menunjukkan bahwa tidak ada outlier dalam data "Happiness Score" ini. Dengan demikian, bisa disimpulkan bahwa **persebaran skor kebahagiaan cukup merata dan tidak ada nilai yang sangat ekstrem atau menyimpang dari mayoritas data.**

```python
import numpy as np

Q1 = df["Happiness Score"].quantile(0.25)
Q3 = df["Happiness Score"].quantile(0.75)
IQR = Q3 - Q1

# Batas Outlier
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(f"Batas Bawah: {lower_bound}")
print(f"Batas Atas: {upper_bound}")
```

> **Analisis Output Outliers**

* IQR dihitung dari **selisih antara kuartil ketiga (Q3) dan kuartil pertama (Q1)**, kemudian digunakan untuk menentukan batas bawah dan batas atas dengan rumus Q1 - 1.5 × IQR dan Q3 + 1.5 × IQR.

* Nilai yang **berada di luar rentang tersebut dianggap sebagai outlier atau data pencilan,** karena tidak mencerminkan pola umum dalam distribusi data. Dengan kata lain, negara-negara yang memiliki skor kebahagiaan di bawah **2.462 atau di atas 8.111 dianggap sebagai pengecualian dalam pola sebaran skor kebahagiaan secara keseluruhan.**

```python
# Menampilkan Outliers
outliers = df[(df["Happiness Score"] < lower_bound) | (df["Happiness Score"] > upper_bound)]
print(outliers)
```
